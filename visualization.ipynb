```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("train.csv")
display(data.head(10))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['PassengerId'], data['Age'],color="black")
plt.title("Scatter Plot")
plt.xlabel('PassengerId')
plt.ylabel('Age')
plt.colorbar()
plt.show()
```

```python
plt.plot(data['Survived'])
plt.plot(data['Parch'])
plt.title('Line Chart')
plt.xlabel('Survived')
plt.ylabel('Parch')
 
plt.show()
```

```python
import seaborn as sns
sns.histplot(x='Age',data=data,color="green")
plt.show()
```

```python
sns.lineplot(x='Parch', y='Pclass', data=data)
plt.show()
```

```python
sns.boxplot(x='Pclass',y='Age', data=data)
 
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=data['PassengerId']
y=data['Parch']
z=data['Survived']
ax.scatter(x,y,z,color="brown")
plt.show()
```
